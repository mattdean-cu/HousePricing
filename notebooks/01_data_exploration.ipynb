# Data Exploration for Ames Housing Dataset
Goals to explore:
 - understanding the structure of the data
 - identify missing values
 - Explore distributions and relationships
 - Discover the best features that correlate to house prices

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/train.csv')
print(f"Shape: {df.shape}")
df.head()
```

## Dataset info
-checking general info about columns and datatypes

```python
df.info()
```

```python
df.describe()
```

## Missing Values

-Finding Columns with missing data

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing
```

## Sale Price Distribution
-examining distributions of target variables

```python
sns.histplot(df['SalePrice'], kde=True)
plt.title("SalePrice Distribution")
plt.show()
```

## Feature Correlation with Sales Prices of Houses
-looking at features that correlate the most with the sale price

```python
corr = df.corr(numeric_only=True)
top_corr = corr['SalePrice'].abs().sort_values(ascending=False).head(10).index

sns.heatmap(df[top_corr].corr(), annot=True, cmap='coolwarm')
plt.title("Top 10 Correlated Features")
plt.show()
```

## Outlier Check (living area vs price)

```python
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df,
    x='GrLivArea',
    y='SalePrice',
    hue='OverallQual',
    palette='viridis',
    alpha=0.7
)

plt.title('GrLivArea vs SalePrice Colored by OverallQual')
plt.xlabel('Above Ground Living Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.legend(title='Overall Quality', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
